```python
def rgb2hex(r, g, b):
    rgb = [r, g, b]
    answ = ''
    for c in rgb:
        if len(hex(c)[2:]) > 1:
            answ += hex(c)[2:]
        else:
            answ += '0' + hex(c)[2:] 
    return answ
```
